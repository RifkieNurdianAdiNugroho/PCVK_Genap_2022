```python
import matplotlib.pyplot as plt
```

```python
flavors = ('Chocolate','Vanilla','Pistachio','Mango','Strawberry')
votes = (12,11,4,8,7)
colors = ('#8B4513','#FFF8DC','#93C572','#E67F0D','#D53032')
explode =(0,0,0,0.1,0)
```

```python
plt.pie(
    votes,
    labels=flavors,
)
plt.show()
```

```python
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
)
plt.show()
```

```python
plt.pie(
    votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
)
plt.show()
```
